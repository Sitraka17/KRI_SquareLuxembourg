```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for Investment Grade % of weighted average PD
quarter_labels = ["Q1", "Q2", "Q3", "Q4"]
inv_grade_pd = [1.20, 1.30, 1.40, 1.50]

# Set up the figure and axes
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# Plot the bar chart for Investment grade share %
inv_grade_share = [30, 40, 50, 60]
ax1.bar(quarter_labels, inv_grade_share, color="blue")
ax1.set_ylabel("Investment grade share (%)")
ax1.set_ylim(0, 60)

# Plot the bar chart for Investment Grade % of weighted average PD
ax2.bar(quarter_labels, inv_grade_pd, color="green")
ax2.set_ylabel("Probability of Default (%)")
ax2.set_ylim(0, 2)

# Set the title and x-axis label
ax1.set_title("Investment Grade % of weighted average PD")
ax1.set_xlabel("Quarters of 2022")

plt.show()
```

```python
import random
import matplotlib.pyplot as plt

low_risk = list(range(2, 11))   # x-axis values for low risk transactions
high_risk = list(range(10, 23)) # x-axis values for high risk transactions
x_values = low_risk + high_risk

y_values = [random.randint(10, 500) for _ in range(len(x_values))]

fig, ax = plt.subplots()

ax.bar(low_risk, y_values[:len(low_risk)], label="Low Risk")
ax.bar(high_risk, y_values[len(low_risk):], label="High Risk")

ax.set_xlabel("Type of Transaction")
ax.set_ylabel("Random Values")

ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the data for the pie chart
currencies = ['€', '$', '¥']
values = [75, 10, 15]
# Set up the figure and plot the pie chart
fig, ax = plt.subplots()
ax.pie(values, labels=currencies)

# Set the title
ax.set_title("Breakdown of currency")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the bar chart
categories = ['Low Risk', 'High Risk']
stages = ['1st Qtl', '2nd Qtl', '3rd Qtl']
data = np.random.rand(len(categories), len(stages))

# Set up the figure and plot the bar chart
fig, ax = plt.subplots()
bar_width = 0.35
bar_positions = np.arange(len(stages))
for i, category in enumerate(categories):
    ax.bar(bar_positions + i * bar_width, data[i], bar_width, label=category)

# Set the title and axis labels
ax.set_title('Breakdown by asset classification type and IFRS 9 stage')
ax.set_ylabel('Performing')
ax.set_yticks([0, 1])
ax.set_yticklabels(categories)
ax.set_xlabel('% ')
ax.set_xticks(bar_positions + bar_width / 2)
ax.set_xticklabels(stages)

# Add a legend
ax.legend()

plt.show()
```
